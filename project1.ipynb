```python
import pandas as pd
import numpy as np
df = pd.read_csv('uiuc-students-by-curriculum.csv')
```

```python
df
```

```python
df = df[df["Major Name"].notnull()]
df = df[df["Major Name"] != "UNASSIGNED"]
df = df[df["Total"] != "Art and Design"]
null_degrees = df[df["Degree"].isnull()]
df = df[df["Degree"].notnull()]
df = df[df["Degree"].str.startswith("B")]

df = pd.concat([null_degrees, df])
```

```python
del df["Term Code"]
del df["Department"]
del df["Degree"]
del df["Major Code"]
del df["Concentration Code"]
del df["Concentration Name"]
del df["Unknown"]
del df["Caucasian"]
del df["Asian American"]
del df["African American"]
del df["Hispanic"]
del df["Native American"]
del df["Hawaiian/Pacific Isl"]
del df["Multiracial"]
del df["International"]
del df["Unknown.1"]
del df["All African American"]
del df["All Native American"]
del df["All Hawaiian/ Pac Isl"]
del df["All Asian"]
del df["Academic Program Code"]
del df["Legacy Advising Code"]
del df["Legacy Department Code"]
del df["Legacy Major Name"]
```

```python
def convert_str(num_str):
    try:
        return int(num_str.replace(',', '').replace(' ', ''))
    except (TypeError, AttributeError):
        return num_str

df['Total'] = df['Total'].apply(convert_str)
df['Male'] = df['Male'].apply(convert_str)
df['Female'] = df['Female'].apply(convert_str)
df['Illinois'] = df['Non-Illinois'].apply(convert_str)
```

```python
aggregation_functions = {'Total': 'sum', 'Male': 'sum', 'Female': 'sum', 'Illinois': 'sum', 'Non-Illinois': 'sum'}
df = df.groupby(["Fall", "Major Name", "College"]).sum()
```

```python
df.to_csv('uiuc-cleaned.csv')
```
